````python
import scipy.stats as st
````

````python
np.random.seed(42)
````

````python
x = st.skewnorm.rvs(10, size=500)
````

````python
x[0:20]
````

````python
fig, ax = plt.subplots()
_ = plt.hist(x, color = 'lightblue')
````

````python
xbar = x.mean()
xbar
````

````python
fig, ax = plt.subplots()
plt.axvline(x = x.mean(), color='orange')
_ = plt.hist(x, color = 'lightblue')
````

````python
np.median(x)
````

````python
fig, ax = plt.subplots()
plt.axvline(x = np.mean(x), color='orange')
plt.axvline(x = np.median(x), color='green')
_ = plt.hist(x, color = 'lightblue')
````

````python
sigma = x.std()
sigma
````

````python
fig, ax = plt.subplots()
plt.axvline(x = xbar, color='orange')
plt.axvline(x = xbar+sigma, color='olivedrab')
plt.axvline(x = xbar-sigma, color='olivedrab')
_ = plt.hist(x, color = 'lightblue')
````
